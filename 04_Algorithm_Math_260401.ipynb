#### 약수 구하기

```python
# 숫자 N의 약수 구하는 알고리즘
divisor_list = [] # 빈 리스트 만들기
N = 12 # N은 12
for num in range(1, N + 1): # 1부터 N까지
    if N % num == 0: # 12를 1부터 나누어 떨어지는지 확인
        divisor_list.append(num) # 나누어 떨어지면 (약수이면) 리스트에 추가

print(divisor_list)
```

#### 공약수 구하기

```python
#기준 숫자 정하기
def swap(a, b): # swap 함수 만들기
    a, b = b, a
    return a, b
    
M, N = 12, 15 # M = 12, N = 15

if M > N: swap(M, N) # M이 N보다 크면 둘이 바꿈

# M의 약수를 찾으면 그 약수가 N의 약수인지 확인
cbivisor_list = [] # 빈 리스트 생성
for num in range(1, M + 1): # 1부터 M까지
    if M % num == 0 and N % num == 0: # M과 N 둘다 나누어 떨어지면
        cbivisor_list.append(num) # 리스트에 추가

print(cbivisor_list)
```

#### 최대공약수 구하기

```python
M, N = 20, 10 # M = 20, N = 10

# 기준 숫자 정하기
if M > N: swap(M, N) # 위에 함수 만들어놔서 여기서도 됨

# M의 약수를 큰 수부터 구하여, N의 약수가 되는 수는 찾으면 몀춤
for num in range(M, 0, -1): # M부터 1까지 1씩 감소
    if (M % num == 0) and (N % num == 0): # M, N 둘다 나누어 떨어지면
        gcd = num # 큰수부터 1까지 두 수 모두 나누어 떨어뜨리면 그게 최대 공약수
        break
        
print(gcd)
```

#### 최소공배수 구하기

```python
M, N = 7, 3

if M > N: swap(M,N)

i = 1 # 1부터 (0은 곱하면 안됨)

while True:
    multiple = N * i # 큰 수에 1부터 곱해가며
    if multiple % M == 0: # 그 수가 작은수의 배수라면
        lcm = multiple # 그게 최소공배수
        break
    i += 1 # 배수가 아니라면 1증가해서 다시 비교

print(lcm)
```

#### 소수 (Prime Number)

```python
def is_prime_number(N):
    if N == 1: return False # 1은 소수가 아니므로 바로 False

    is_prime = True # 기본값으로 True
    for i in range(2, N): # 1과 자기 자신을 제외한 수로 나누어 떨어지면 안되기 때문에 2부터 N-1까지
        if N % i == 0: # 나누어 떨어지면 
            is_prime = False # False로 바꾸고 반복 멈춤
            break

    return is_prime # 최종적으로 저장된 is_prime 반환
    
print(is_prime_number(7))
```

```python
def is_prime_number(N):
    if N == 2: return True # 2는 소수이므로 True
    if N ==1 or N % 2 == 0: return False # 1이거나 2의 배수라면 False

    for i in range(3, N, 2): # 2의 배수는 걸러졌으므로 3부터 2씩 뛰어 홀수만 비교
        if N % i == 0: # 나누어 떨어지면 
            return False # False
            
    return True # 안걸리고 탈출하면 소수이므로 True
    
print(is_prime_number(7))
```

#### 투 포인터 연습 (문제)

```python
num_list = [1, 1, 1, 1]
N, M = 4, 2
answer = 0 
start = end = 0 # 시작, 끝 포인터 모두 0부터 시작
sum = num_list[0] # 초기값은 첫번째 값으로 시작
while end < N: # 끝 포인터가 배열 끝까지 도달하기 전까지 반복
    if sum == M: answer += 1 # sum이 목표값과 같아지면 answer 1 카운트
    if sum <= M: # 위에서 안걸리고 목표값보다 작거나 같으면 끝포인터 1 증가
        end += 1
        if end < N: sum += num_list[end] # 끝포인터가 배열 끝까지 도달하지 않았을때만 끝포인터 값 더함
    else: # 목표값보다 커지면 시작 포인터 값 빼고 시작포인터 1 증가
        sum -= num_list[start] 
        start += 1
print(answer)
```

#### 투 포인터 연습 (BOJ문제)

```python
A_list = [2, 3, 5, 9]
B_list = [1, 4, 7]
merged_list = []

pA = pB = 0

N, M = len(A_list), len(B_list)

while pA < N and pB < M: # 두 포인터 모두 배열 수 보다 작을때 반복
    if A_list[pA] < B_list[pB]: # A리스트의 A포인터 값보다 B가 더 크면 A값을 저장하고 A포인터는 1 증가
        merged_list.append(A_list[pA])
        pA += 1
    else: # 반대라면 B저장 후 B포인터 1 증가
        merged_list.append(B_list[pB])
        pB += 1
# 둘 중 한 포인터는 먼저 배열 크기를 넘어가기 떄문에 남은 값들을 저장해줘야함
while pA < N: # B가 먼저 나갔을 때 A값 저장
    merged_list.append(A_list[pA])
    pA += 1
while pB < M: # A가 먼저 나갔을 때 B값 저장
    merged_list.append(B_list[pB])
    pB += 1

print(merged_list)
```

```python
# 정수 N을 받아서 base에 댛나 진법의 수로 변환
def convert_base(N, base):
    # N을 base로 나눈 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    # 나눈 나머지를 진법에 해당하는 숫자로 변환하여 리스트에 저장한다.
    # N : 몫을 저장
    # new_list : 나머지를 저장하는 리스트
    new_number = ""
    digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    while N != 0:
        new_number = digits[N % base] + new_number
        N //= base
        print(N)
    return new_number

print(convert_base(25,16))
```
