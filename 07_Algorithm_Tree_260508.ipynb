### Binary Search Tree

```python
class STreeNode:
    def __init__(self, key, left, right):
        self.key = key
        self.left = left
        self.right = right
```

```python
# BST 생성
t3 = STreeNode(3, None, None) #                  t18
t12 = STreeNode(12, None, None) #            /          \
t7 = STreeNode(7, t3, t12) #             t7               t26
t27 = STreeNode(27, None, None) #     /      \          /      \
t31 = STreeNode(31, t27, None) #    t3       t12     None      t31
t26 = STreeNode(26, None, t31) #   /  \     /   \             /   \
t18 = STreeNode(18, t7, t26) #  None  None N     N          t27   None
my_bst = t18 #                                             N   N                                               
```

```python
def search_bst(n, key):
    if n == None: return None
    if key == n.key: return n
    if key < n.key:
        return search_bst(n.left, key)
    else:
        return search_bst(n.right, key)
    
result = search_bst(my_bst, 25)
if result: print(result.key)
else:
    print('없는 key입니다.')
```

```python
def search_bst_iter(n, key):
    while n != None:
        if key == n.key: return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst, 31)
if result: print(result.key)
else:
    print('없는 key입니다.')
```

```python
bst = STreeNode(None, None, None) # 빈 노드 생성(현재 노드), 이 노드에 받은 숫자 추가할거임

def insert_bst(num): # 추가할 숫자 받기
    if bst.key == None: # 현재 노드에 아무 값이 없으면
        bst.key = num   # 받은 숫자가 바로 키가 됨, 맨 처음만 작동
        return # 추가 됬으니 내보내기
    number = STreeNode(num, None, None) # 받은 숫자를 키로 하는 새 노드 만들기, 노드로 안만들면 다음에 추가할 곳이 없음
    root = bst # 함수 내에서 계속 바뀌는 루트 노드가 필요함, 처음엔 현재 노드를 가리킴(커서, current)
    insert = None # 바뀐 루트 노드를 저장할 곳이 필요함, 마지막에 이 노드 왼,오른쪽에 받은 숫자를 추가할거임(부모 노드)
    while root != None: # 루트가 None이라면 멈추고 숫자 추가, 아니면 반복
        if num < root.key: # 받은 숫자가 지금 보고있는 노드(루트 노드)의 키보다 작으면 
            insert = root  # 마지막 루트 노드 저장하고
            root = root.left # 루트를 현재 노드의 왼쪽에 저장된 노드로 변경
        elif num > root.key: # 받은 숫자가 크면
            insert = root # 마지막 루트 노드 저장하고
            root = root.right # 루트는 오른쪽 노드로 변경
        else: # 받은 숫자가 현재 노드 키값과 같으면
            return # 추가할 거 없으니 내보내기
    # 바뀐 루트 노드가 노드가 아니라 None값 이라면 while 나와서 실행
    if num < insert.key: # 마지막으로 저장한 루트 노드의 키보다 작으면
        insert.left = number # 왼쪽
    else:                    # 아니면
        insert.right = number # 오른쪽에 저장
                              # 객체는 내용 다 공유함, 밖에 bst도 같이 바뀜
```
