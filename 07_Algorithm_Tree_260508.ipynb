### 1. 이진 트리 (Binary Tree)

```python
class TreeNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right
    def set_left(self, left):
        self.left = left
    def set_right(self, right):
        self.right = right
```

```python
my_tree = TreeNode('A', None, None) # A 객체 생성
new_node = TreeNode('B', None, None) # B 객체 생성
my_tree.set_left(new_node) # A 객체의 왼쪽에 B 객체 주소 저장
new_node = TreeNode('C', None, None) # C 객체 생성
my_tree.set_right(new_node) # A 객체의 오른쪽에 C 객체 주소 저장

cur_node = my_tree.left # B 객체 주소 저장
cur_node.set_left(TreeNode('D', None, None)) # B 객체의 왼쪽에 D 객체 생성
cur_node.set_right(TreeNode('E', None, None)) # B 객체의 오른쪽에 E 객체 생성

cur_node = my_tree.right
```
