```python
def binary_search_recursive(A, key, low, high): # 데이터와 찾을 값을 받고 범위 인덱스값을 받음
    # 탈출조건
    if (low > high): # 시작값이 끝값보다 커지면 -1 반환(찾는 값이 없음)
        return -1
    # 중앙값
    middle = (low + high) // 2 # 더하는 이유는 시작값부터 끝값까지의 범위에서 중간을 찾기 위함
                               # 0 ~ 15 > 15//2 = 7, 7(0) ~ 22(15) > 22//2 = 11
    
    if key == A[middle]: return middle # 찾는값이 A의 중앙값과 같으면 그대로 중앙값(찾는값) 반환
        
    elif key < A[middle]: # 찾는값이 중앙값보다 작으면 끝값을 중앙값 왼쪽으로 당김(위에서 중앙값과 같은지 비교했기 때문에 중앙값-1)
        return binary_search_recursive(A, key, low, middle-1) # 끝값에 바뀐 끝값을 넣고 다시
        
    else: # 찾는값이 중앙값보다 크면 시작값을 중앙값 오른쪽으로 옮김
        return binary_search_recursive(A, key, middle+1, high) # 시작값만 바꿔서 다시
        
# 중앙값이 찾는값과 같지 않다면 시작값이나 끝값을 옮기는데 반드시 시작값과 끝값이 곂치게 됨
# 곂친다면 시작값과 끝값을 더하고 2로 나눠도(중앙값) 자기 자신이기 때문에 중앙값 = 시작값 = 끝값
# 이때 찾는값과 중앙값이 같지 않다면 if 넘김 > elif 넘김 > else에서 시작값을 1올림 > 탈출조건에서 걸림
# elif에 중앙값보다 큰지를 먼저 비교했다면 탈출조건도 바꿔야함
```

```python
def binary_search(A,key): # 데이터와 찾을값을 받음
    low = 0
    high = len(A)-1 # 시작 인덱스와 끝 인덱스를 정함
    
    while low <= high: # 시작값이 끝값보다 커지면 (찾는값 없으면) 멈춤
        middle = (low + high) // 2 # 중앙값
        if A[middle] == key: # 중앙값이 찾는값이면 그대로 반환
            return middle
        elif A[middle] > key: # 중앙값이 찾는값보다 크면 끝값을 옮기고 다시
            high = middle-1
        else: # 중앙값이 찾는값보다 작으면 시작값 옮기고 다시
            low = middle+1
    return -1 # 반환 못하고 나오면(찾는값 없으면) -1 반환
```

```python
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]

num = int(input("찾을 숫자 : "))

index = binary_search_recursive(data, num, 0, len(data)-1) # 찾는값의 인덱스 받기
if index == -1: # 반환값이 -1이면(찾는값이 없으면)
    print("없는 숫자입니다.")
else: # 반환값 있으면(찾음)
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.') # 찾는값 인덱스 출력

index = binary_search(data, num) # 찾는값 인덱스 받기
if index == -1:
    print("없는 숫자입니다.")
else:
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.')
```
